```python
from transformers import AutoTokenizer
from transformers import AutoModelForCausalLM

MODEL_NAME="Qwen/Qwen2.5-0.5B-Instruct"

tokenizer=AutoTokenizer.from_pretrained(MODEL_NAME)

model=AutoModelForCausalLM.from_pretrained(
    MODEL_NAME
)
```

```python
#Load PDFs#
```

```python
import os
from pypdf import PdfReader

pdf_folder = r"C:\Users\sampath\Documents\Assignment02\data"

documents = []

for file in os.listdir(pdf_folder):

    if file.endswith(".pdf"):

        reader = PdfReader(os.path.join(pdf_folder, file))

        text = ""

        for page in reader.pages:
            page_text = page.extract_text()

            if page_text:
                text += page_text

        documents.append(text)

print(f"Loaded {len(documents)} PDF(s)")
```

```python
#Clean text#
```

```python
lemmatizer = WordNetLemmatizer()

stop_words = set(stopwords.words("english"))
```

```python
import re
```

```python
def preprocess(text):

    text = text.lower()

    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z ]', ' ', text)

    words = word_tokenize(text)

    words = [lemmatizer.lemmatize(word) for word in words]

    words = [word for word in words if word not in stop_words]

    return " ".join(words)
```

```python
documents = [preprocess(doc) for doc in documents]

print("Preprocessing completed!")
print("Total documents:", len(documents))
```

```python
def split_text(text,
               chunk_size=500,
               overlap=100):

    chunks=[]

    start=0

    while start < len(text):

        end=start+chunk_size

        chunks.append(text[start:end])

        start=end-overlap

    return chunks
```

```python
chunks=[]

for doc in documents:

    chunks.extend(split_text(doc))
```

```python
from sentence_transformers import SentenceTransformer
```

```python
embedding_model = SentenceTransformer(
    "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
print(len(chunks))
```

```python
embeddings = embedding_model.encode(chunks)

print(embeddings.shape)
```

```python
import numpy as np
import faiss

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings).astype("float32"))

print("Total vectors:", index.ntotal)
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))
```

```python
def retrieve(question,k=3):

    q_embedding = embedding_model.encode([question])

    distance,indexes = index.search(np.array(q_embedding),k)

    context=[]

    for i in indexes[0]:

        context.append(chunks[i])

    return "\n".join(context)
```

```python
def build_prompt(question):

    context = retrieve(question)

    prompt = f"""

You are SeaCareAI.

Answer ONLY from the context.

If the answer is not available,
reply

"I couldn't find this information in the uploaded maritime documents."

Context

{context}

Question

{question}

Answer

"""

    return prompt
```

```python
def ask_llm(question):

    prompt = build_prompt(question)

    messages = [
        {"role":"user","content":prompt}
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer(text,
                       return_tensors="pt").to(model.device)

    output = model.generate(
        **inputs,
        max_new_tokens=200,
        temperature=0.3
    )

    answer = tokenizer.decode(
        output[0],
        skip_special_tokens=True
    )

    return answer
```

```python
MARITIME_KEYWORDS = [

"ship",

"vessel",

"captain",

"engine",

"seafarer",

"imo",

"marlpol",

"solas",

"stcw",

"bridge",

"navigation",

"cargo",

"safety"

]
```

```python
def is_domain_question(question):

    q = question.lower()

    for word in MARITIME_KEYWORDS:

        if word in q:

            return True

    return False
```

```python
BAD_WORDS = [

"hack",

"virus",

"bomb",

"kill",

"weapon"

]
```

```python
def safe_question(question):

    q = question.lower()

    for word in BAD_WORDS:

        if word in q:

            return False

    return True
```

```python
import time

def ask_seacare(question, return_latency=False):

    # Domain safeguard
    if not is_domain_question(question):
        if return_latency:
            return (
                "This question is outside the scope of the uploaded maritime documents.",
                0.0
            )
        return "This question is outside the scope of the uploaded maritime documents."

    start = time.time()

    answer = ask_llm(question)

    latency = time.time() - start

    if return_latency:
        return answer, latency

    return answer
```

```python
result = ask_seacare("What is SOLAS?", return_latency=True)

print(result)
print(type(result))
```

```python
questions = [
    "What is seasickness?",
    "What is SOLAS?",
    "How should a fire be handled on board?"
]

times = []

print("=== SeaCareAI Evaluation ===\n")

for q in questions:

    answer, latency = ask_seacare(q, return_latency=True)

    times.append(latency)

    print(f"Question : {q}")
    print(f"Latency : {latency:.2f} seconds")
    print(f"Answer   : {answer[:100]}...")
    print("-" * 60)

average = sum(times) / len(times)

print(f"\nAverage Latency : {average:.2f} seconds")
```
